```python
from main import remove_rain_streaks
import matplotlib.pyplot as plt
```

```python
O, B, R = remove_rain_streaks("data/case1.jpg")

_, axs = plt.subplots(1, 2, figsize=(7, 5))
axs[0].imshow(O)
axs[0].set_title("Original picture")
axs[0].axis("off")

axs[1].imshow(B)
axs[1].set_title("Processed picture")
axs[1].axis("off")
plt.tight_layout()
plt.show()
```
